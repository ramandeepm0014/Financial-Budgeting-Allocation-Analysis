```python
!pip install plotly -q

import plotly.graph_objects as go
import plotly.express as px
import pandas as pd
```

```python
df = pd.read_csv("Financial_Budgeting_Dataset.csv")
df.head()
```

```python
df = df.dropna()
df.info()
```

```python
fig = px.sunburst(
    df,
    path=[
        "Department",
        "Expense_Category",
        "Budget_Status"
    ],
    values="Budget_Allocated",
    color="Allocation_Efficiency",
    color_continuous_scale="Viridis",
    title="Budget Allocation Hierarchy"
)

fig.update_layout(
    template="plotly_dark",
    height=750,
    title_x=0.5
)

fig.show()
```

```python
sankey_data = (
    df.groupby(
        ["Department","Expense_Category"]
    )["Budget_Allocated"]
    .sum()
    .reset_index()
)

labels = list(
    pd.concat([
        sankey_data["Department"],
        sankey_data["Expense_Category"]
    ]).unique()
)

source = [
    labels.index(i)
    for i in sankey_data["Department"]
]

target = [
    labels.index(i)
    for i in sankey_data["Expense_Category"]
]

value = sankey_data["Budget_Allocated"]

fig = go.Figure(
    data=[
        go.Sankey(
            node=dict(
                pad=20,
                thickness=20,
                label=labels
            ),
            link=dict(
                source=source,
                target=target,
                value=value
            )
        )
    ]
)

fig.update_layout(
    title="Budget Flow Analysis",
    template="plotly_dark",
    height=700
)

fig.show()
```

```python
fig = px.treemap(
    df,
    path=[
        "Department",
        "Expense_Category"
    ],
    values="Budget_Allocated",
    color="Budget_Variance",
    title="Budget Distribution Treemap"
)

fig.update_layout(
    template="plotly_dark",
    height=700
)

fig.show()
```

```python
top_dept = (
    df.groupby("Department")
    ["Budget_Allocated"]
    .sum()
    .sort_values(ascending=False)
    .reset_index()
)

fig = px.bar(
    top_dept,
    x="Department",
    y="Budget_Allocated",
    color="Budget_Allocated",
    text_auto=".2s",
    title="Department Wise Budget Allocation"
)

fig.update_layout(
    template="plotly_dark",
    height=600
)

fig.show()
```

```python
fig = px.scatter(
    df,
    x="Budget_Allocated",
    y="Budget_Utilized",
    color="Department",
    size="Monthly_Expense",
    hover_data=[
        "Expense_Category",
        "Budget_Status"
    ],
    title="Budget Allocation vs Utilization"
)

fig.update_layout(
    template="plotly_dark",
    height=700
)

fig.show()
```

```python
quarter = (
    df.groupby("Fiscal_Quarter")
    ["Budget_Allocated"]
    .sum()
    .reset_index()
)

fig = px.line(
    quarter,
    x="Fiscal_Quarter",
    y="Budget_Allocated",
    markers=True,
    title="Quarterly Budget Trend"
)

fig.update_layout(
    template="plotly_dark",
    height=650
)

fig.show()
```

```python
status = (
    df["Budget_Status"]
    .value_counts()
    .reset_index()
)

status.columns = [
    "Status",
    "Count"
]

fig = px.pie(
    status,
    names="Status",
    values="Count",
    title="Budget Status Distribution"
)

fig.update_layout(
    template="plotly_dark",
    height=650
)

fig.show()
```

```python
print(
    "Total Budget Allocated:",
    round(df["Budget_Allocated"].sum(),2)
)

print(
    "Total Budget Utilized:",
    round(df["Budget_Utilized"].sum(),2)
)

print(
    "Total Revenue:",
    round(df["Actual_Revenue"].sum(),2)
)

print(
    "Average Allocation Efficiency:",
    round(df["Allocation_Efficiency"].mean(),2)
)
```
